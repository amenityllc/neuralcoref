```python
import pandas as pd
import json
import ipywidgets as widgets
```

```python
with open('runs/scores.txt') as f:
    records = [json.loads(line) for line in f]
```

```python
df_scores = pd.DataFrame.from_records(records).drop('test_file', axis=1)
```

```python
df_scores
```

```python
dropdown_dataset = widgets.Dropdown(options =['ALL', 'dev', 'minidev', 'train'])

def dropdown_dataset_eventhandler(change):
    if (change.new == 'ALL'):
        display(df_scores)
    else:
        display(df_scores[df_scores['dataset'] == change.new])


dropdown_dataset.observe(dropdown_dataset_eventhandler, names='value')

display(dropdown_dataset)
```

```python
df_scores.drop('timestamp', axis=1, inplace=True)
df_scores.sort_values('muc_precision', inplace=True)
```

```python
df_scores = df_scores[df_scores['args'].notnull()]
```
